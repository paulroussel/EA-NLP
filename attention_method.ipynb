```python
from models import *

from util import Dictionary, get_args

import torch
import torch.nn as nn
import torch.optim as optim
#from torch.autograd import Variable

import json
import time
import random
import os
```

```python
def Frobenius(mat):
    size = mat.size()
    if len(size) == 3:  # batched matrix
        ret = (torch.sum(torch.sum((mat ** 2), 1), 1).squeeze() + 1e-10) ** 0.5
        return torch.sum(ret) / size[0]
    else:
        raise Exception('matrix for computing Frobenius norm should be with 3 dims')


def package(data):
    """Package data for training / evaluation."""
    data = list(map(lambda x: json.loads(x), data))
    dat = list(map(lambda x: list(map(lambda y: dictionary.word2idx[y], x['text'])), data))
    maxlen = 0
    for item in dat:
        maxlen = max(maxlen, len(item))
    targets = list(map(lambda x: x['label'], data))
    maxlen = min(maxlen, 500)
    for i in range(len(data)):
        if maxlen < len(dat[i]):
            dat[i] = dat[i][:maxlen]
        else:
            for j in range(maxlen - len(dat[i])):
                dat[i].append(dictionary.word2idx['<pad>'])
    dat = torch.LongTensor(dat)
    targets = torch.LongTensor(targets)
    return dat.t(), targets

@torch.no_grad()
def evaluate():
    """evaluate the model while training"""
    model.eval()  # turn on the eval() switch to disable dropout
    total_loss = 0
    total_correct = 0
    for batch, i in enumerate(range(0, len(data_val), batch_size)):
        data, targets = package(data_val[i:min(len(data_val), i+batch_size)])
        if cuda_arg:
            data = data.cuda()
            targets = targets.cuda()
        hidden = model.init_hidden(data.size(1))
        output, attention, _ = model.forward(data, hidden)
        output_flat = output.view(data.size(1), -1)
        total_loss += criterion(output_flat, targets).data
        #print(output_flat)
        prediction = torch.max(output_flat, 1)[1]
        #print(prediction)
        total_correct += torch.sum((prediction == targets).float())
    return total_loss.item() / (len(data_val) // batch_size), total_correct.data.item() / len(data_val)


def train(epoch_number):
    global best_val_loss, best_acc
    model.train()
    total_loss = 0
    total_pure_loss = 0  # without the penalization term
    start_time = time.time()
    for batch, i in enumerate(range(0, len(data_train), batch_size)):
        
        #print("len:"+str(len(data_train)))
        #print("(batch, i): "+'('+str(batch)+','+str(i)+')')
        start_time_epoch = time.time()
        data, targets = package(data_train[i:i+batch_size])
        if cuda_arg:
            data = data.cuda()
            targets = targets.cuda()
        hidden = model.init_hidden(data.size(1))
        output, attention, out_bool = model.forward(data, hidden)
        loss = criterion(output.view(data.size(1), -1), targets)
        total_pure_loss += loss.data.item()
        if out_bool:  # add penalization term
            attentionT = torch.transpose(attention, 1, 2).contiguous()
            extra_loss = Frobenius(torch.bmm(attention, attentionT) - I[:attention.size(0)])
            loss += penalization_coeff * extra_loss
        optimizer.zero_grad()
        loss.backward()

        nn.utils.clip_grad_norm_(model.parameters(), clip)
        optimizer.step()

        total_loss += loss.data.item()

        if batch % log_interval == 0 and batch > 0:
            elapsed = time.time() - start_time
            print('| epoch {:3d} | {:5d}/{:5d} batches | ms/batch {:5.2f} | loss {:5.4f} | pure loss {:5.4f}'.format(
                  epoch_number, batch, len(data_train) // batch_size,
                  elapsed * 1000 / log_interval, total_loss / log_interval,
                  total_pure_loss / log_interval))
            total_loss = 0
            total_pure_loss = 0
            start_time = time.time()

#            for item in model.parameters():
#                print item.size(), torch.sum(item.data ** 2), torch.sum(item.grad ** 2).data[0]
#            print model.encoder.ws2.weight.grad.data
#            exit()
        print((time.time() - start_time_epoch))
    evaluate_start_time = time.time()
    val_loss, acc = evaluate()
    print('-' * 89)
    fmt = '| evaluation | time: {:5.2f}s | valid loss (pure) {:5.4f} | Acc {:8.4f}'
    print(fmt.format((time.time() - evaluate_start_time), val_loss, acc))
    print('-' * 89)
    # Save the model, if the validation loss is the best we've seen so far.
    if not best_val_loss or val_loss < best_val_loss:
        with open(save, 'wb') as f:
            torch.save(model, f)
        f.close()
        best_val_loss = val_loss
    else:  # if loss doesn't go down, divide the learning rate by 5.
        for param_group in optimizer.param_groups:
            param_group['lr'] = param_group['lr'] * 0.2
    if not best_acc or acc > best_acc:
        with open(save[:-3]+'.best_acc.pt', 'wb') as f:
            torch.save(model, f)
        f.close()
        best_acc = acc
    with open(save[:-3]+'.epoch-{:02d}.pt'.format(epoch_number), 'wb') as f:
        torch.save(model, f)
    f.close()
```

```python

train_data = "train_tok_pressmium_content.json"
val_data = "val_tok_pressmium_content.json" 
test_data = "test_tok_pressmium_content.json" 
emsize =300
nhid =300 
nfc = 300 
dropout =0.5 
attention_unit =350 
attention_hops = 1
epochs = 4
lr = 0.01 
clip = 0.5 
dictionary_arg = "dict_file_pressmium_content" 
word_vector = "/Users/rousselpaul/Documents/cours/3A/deep learning/Projet/NLP-attention/embedding_content.txt.pt" 
save = "models/model-medium.pt" 
batch_size = 100
class_number = 11
optimizer_arg = 'Adam'
attention_hops = 4 
penalization_coeff = 1.0 
log_interval = 100
nlayers = 2
cuda_arg = False
```

```python

    # parse the arguments
#args = get_args()

    # Set the random seed manually for reproducibility.
    #torch.manual_seed(args.seed)
    #if torch.cuda.is_available():
    #    if not args.cuda:
    #        print("WARNING: You have a CUDA device, so you should probably run with --cuda")
    #    else:
    #        torch.cuda.manual_seed(args.seed)
    #random.seed(args.seed)

    
    
    # Load Dictionary
    #assert os.path.exists(train_data)
    #assert os.path.exists(val_data)
    
print('Begin to load the dictionary.')
dictionary = Dictionary(path=dictionary_arg)


best_val_loss = None
best_acc = None

n_token = len(dictionary)

model = Classifier({
    'dropout': dropout,
    'ntoken': n_token,
    'nlayers': nlayers,
    'nhid': nhid,
    'ninp': emsize,
    'pooling': 'mean',
    'attention-unit': attention_unit,
    'attention-hops': attention_hops,
    'nfc': nfc,
    'dictionary': dictionary,
    'word-vector': word_vector,
    'class-number': class_number
})

if cuda_arg:
    model = model.cuda()

I = torch.zeros(batch_size, attention_hops, attention_hops)
for i in range(batch_size):
    for j in range(attention_hops):
        I.data[i][j][j] = 1
if cuda_arg:
    I = I.cuda()

criterion = nn.CrossEntropyLoss()
if optimizer_arg == 'Adam':
    optimizer = optim.Adam(model.parameters(), lr=lr, betas=[0.9, 0.999], eps=1e-8, weight_decay=0)
elif optimizer_arg == 'SGD':
    optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=0.01)
else:
    raise Exception('For other optimizers, please add it yourself. '
                    'supported ones are: SGD and Adam.')
print('Begin to load data.')
data_train = open(train_data).readlines()
data_val = open(val_data).readlines()
step_epoch = 0
try:
    for epoch in range(epochs):
        print(step_epoch)
        train(epoch)
        step_epoch = step_epoch +1
except KeyboardInterrupt:
    print('-' * 89)
    print('Exit from training early.')
    data_val = open(test_data).readlines()
    evaluate_start_time = time.time()
    test_loss, acc = evaluate()
    print('-' * 89)
    fmt = '| test | time: {:5.2f}s | test loss (pure) {:5.4f} | Acc {:8.4f}'
    print(fmt.format((time.time() - evaluate_start_time), test_loss, acc))
    print('-' * 89)
    exit(0)
```
